## cohort table prepration

```python
df["InvoiceDate"].min()
```

```python
df["invoice_year"]= df["InvoiceDate"].dt.year
```

```python
df["invoice_month"]= df["InvoiceDate"].dt.month
```

```python
df
```

```python
df["invoice_year"]=df["invoice_year"].astype("str")
```

```python
df["invoice_month"]=df["invoice_month"].astype("str")
```

```python
df["cohort_date"] = df["invoice_year"]+"-"+ df["invoice_month"]
```

```python
df["cohort_date"] =pd.to_datetime(df["cohort_date"])
```

## extract cohort date to help for analysis

```python
df["invoice_year"]=df["invoice_year"].astype("int")
```

```python
df["invoice_month"]=df["invoice_month"].astype("int")
```

```python
df["customer_first_ac"] =df.groupby("CustomerID")["cohort_date"].transform("min")
```

```python
df["year"] =df["customer_first_ac"].dt.year
```

```python
df["month"]=df["customer_first_ac"].dt.month
```

```python
df["year_diff"] = df["invoice_year"]-df["year"]
```

```python
df["month_diff"] = df["invoice_month"]-df["month"]
```

## create cohort index

```python
df["cohort_indexs"] = df["year_diff"]*12 + df["month_diff"]
```

```python
df["cohort_indexs"].unique()
```

```python
df2 =df.groupby(["customer_first_ac","cohort_indexs"])["CustomerID"].nunique().reset_index()
```

```python
df2
```

## make a pivot table to show the result

```python
df3 =df2.pivot_table(index ="customer_first_ac", columns = "cohort_indexs", values = "CustomerID")
```

```python
df3
```

```python
df3.fillna(0)
```

## viualaize the results into a cohort viz

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df3.index = df3.index.strftime("%b %y")
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df3,annot= True,fmt ="g",cmap="BuPu")

plt.show()
```

## make a retention rate visual

```python
df4 = df3.divide(df3.iloc[:,0],axis =0)
```

```python
df4
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df4,annot= True,fmt =".1%",cmap="Greens")

plt.show()
```
